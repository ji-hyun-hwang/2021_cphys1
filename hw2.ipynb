```python

def logistic(r,f) :
    for i in range(10) :
        f = r*f*(1-f)
        print(f)
    return
print("r:4.5,f:0.5 일때")
logistic(4.5,0.5)
print("r:4.5,f:0.51 일때")
logistic(4.5,0.51)
print("r:4.5,f:0.501 일때")
logistic(4.5,0.501)
```

```python
def fibonacci (n) :
    a = 0
    f = [1,1]
    while a < n :
        a += 1
        fn = f[-1]+ f[-2]
        f.append(fn)
    f = f[:n]
    return f

fibonacci(5)



```
