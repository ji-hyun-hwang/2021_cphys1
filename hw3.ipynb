\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt
import math



def f(n) :
    b = 0
    x = []
    a = [(1/(2*i-1))*(-1)**(i+1) for i in range(1,n,1)]
    for i in a :
        b = b + i
        x.append(b)
    plt.plot(range(1,100,1),x,label = 'Leibniz formula for $\pi$')
    plt.plot([1,100],[np.pi/4,np.pi/4], label = '$\pi$/4')
    plt.legend()
    plt.xlabel("n")
    return
            
f(100)

```

$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$ 

```python
import numpy as np
import matplotlib.pyplot as plt
import math


def f(x,n) :
    i,s = 0,0
    while (i<=n) :
        s = s + ((-1)**i)*(x**(2*i))/math.factorial(2*i+1)
        i += 1
    return s

x = np.arange(-20,20,0.1)

plt.plot(x,f(x,30))
plt.xlabel("x")

plt.plot(x,np.sin(x)/x+1)
```
