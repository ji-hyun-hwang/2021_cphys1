# 1. Leibniz Pi formula

\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots 
\end{align}
이 식은 라이프니츠의 파이 공식이다.  
1. 다음의 식을 구현하는 함수를 만들어라.  
\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}
2. 이 식을 이용하여 아래처럼 그림을 그려라.  

```python
import numpy as np
import matplotlib.pyplot as plt
import math



def f(n) :
    b = 0
    x = []
    a = [(1/(2*i-1))*(-1)**(i+1) for i in np.arange(1,n,1)]
    for i in a :
        b = b + i
        x.append(b)

    b = 0
    y = []
    j = [(1/(2*i-1))*(-1)**(i+1) for i in np.arange(1,n,1)]
    for i in j :
        b = b + i
        y.append(b)
    if n % 2 == 1 : z = int(n/2)
    else : z = int(n/2-1)

    y = y[:z]
    
    plt.figure(figsize=(14,4))
    plt.subplot(1,2,1)
    plt.plot(np.arange(1,n,1),x,label = 'Leibniz formula for $\pi$')
    plt.plot([1,n],[np.pi/4,np.pi/4], label = '$\pi$/4')
    plt.legend()
    plt.xlabel("n")

    

    plt.subplot(1,2,2)
    plt.plot([i for i in np.arange(1,n,1) if i % 2 == 0],y,label = 'Leibniz formula for $\pi$')
    plt.plot([1,n],[np.pi/4,np.pi/4], label = '$\pi$/4')
    plt.legend()
    plt.xlabel("n")


    return
            
f(100)

```

$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$ 

```python
import numpy as np
import matplotlib.pyplot as plt
import math


def f(x,n) :
    i,s = 0,0
    while (i<=n) :
        s = s + ((-1)**i)*(x**(2*i))/math.factorial(2*i+1)
        i += 1
    return s

x = np.arange(-20,20,0.1)
plt.subplot(311)
plt.plot(x,f(x,5), label = "n = 5", color = "r")
plt.plot(x,np.sin(x)/x, "--",color = "b")
plt.ylim(-1,1.5)
plt.xlabel("x")
plt.ylabel("Sinc(x)/x")
plt.legend()

plt.subplot(312)
plt.plot(x,f(x,15),label = "n = 15", color = "r")
plt.plot(x,np.sin(x)/x,"--", color = "b")
plt.ylim(-1,1.5)
plt.xlabel("x")
plt.ylabel("Sinc(x)/x")
plt.legend()

plt.subplot(313)
plt.plot(x,f(x,25),label = "n = 25", color = "r")
plt.plot(x,np.sin(x)/x,"--",color = "b")
plt.ylim(-1,1.5)
plt.legend()
plt.xlabel("x")
plt.ylabel("Sinc(x)/x")
```
