# 1. Error Function

$$ erf(x) = \frac{2}{\sqrt{\pi}} \int_0^x e^{-t^2} dt. $$

* 위와 같이 주어지는 에러함수를 수치적분을 이용하여 구하고 그래프를 그려라. (어떤 방법을 이용하든 관계없음)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import math
```

```python
def f(t) :return np.exp(-(t**2))

x = np.arange(-2.5,2.5,0.01)
z = []
for i in x :
    F,err = integrate.quad(f,0,i)
    k = (2/np.pi**(1/2))*F
    z.append(k)
plt.title("Error Function")
plt.plot(x,z,color = "r")
plt.ylabel("ef(x)")
plt.xlabel("x")
plt.grid()
plt.yticks([-1,-0.5,0,0.5,1])
plt.xticks([-2,-1,0,1,2])
```

$$
\begin{align}
erf(x) &= \frac{2}{\pi} \sum_{n=0}^{\infty} \frac{ (-1)^n x^{2n+1}} { (2n+1) n!}  \\
&= \frac{2}{\pi} \left( x - \frac{x^3}{3} + \frac{x^5}{10} - \frac{x^7}{42} + \cdots \right)
\end{align}
$$

* 에러함수를 테일러 전개하면 위와 같이 주어진다. 위의 전개식과 적분결과를 그래프로 그리고 비교하여라.

```python
x = np.arange(-2.5,2.5,0.01)
def erf(x,n) :
    i,s = 0,0
    while (i<=n) :
        s = s + (2/np.pi**(1/2))*((-1)**i)*x**(2*i+1)/((2*i+1)*math.factorial(i))
        i += 1
    return s
plt.plot(x, erf(x,3),color = "r",label = "n=3")
plt.plot(x, erf(x,5),color = "b",label = "n=5")
plt.plot(x, erf(x,30), color = "g",label = "n=30")
plt.plot(x,z,"--",color = "black",label = "Error Function")
plt.ylim(-1,1)
plt.yticks([-1,-0.5,0,0.5,1])
plt.xticks([-2,-1,0,1,2])
plt.grid()
plt.title("Error Function")
plt.xlabel("x")
plt.ylabel("erf(x)")

```

# 2. (Integration) Trigonometric Function
* sin, cos함수를 trapezoidal 방법을 이용하여 직접 적분하고 그래프를 그려라. 
* sin, cos함수를 scipy.integrate 함수를 이용하여 적분하고 그래프를 그려라.

# Sine

```python
def f(x) : return np.sin(x)
def g(x) : return np.cos(x)

def integ(f,a,b,dx) :
    s,x,k = 0,a,0
    ydata = []
    xdata = []
    while(x<b):
        k = (f(x+dx)+f(x))/2
        s += 0.5*dx*(f(x+dx)+f(x))
        x += dx
        xdata.append(x)
        ydata.append(s)
    plt.title("Sine")
    plt.plot(xdata,ydata, "o", color= "g",label =  "dx =1")
    plt.fill_between(xdata,ydata,alpha=0.2)
    plt.xlabel("x")
    plt.ylabel("Area")
    x = np.arange(a,b,0.1)
    plt.legend()
    return

integ(f,0,9,1)

def integ(f,a,b,dx) :
    s,x,k = 0,a,0
    ydata = []
    xdata = []
    while(x<b):
        k = (f(x+dx)+f(x))/2
        s += 0.5*dx*(f(x+dx)+f(x))
        x += dx
        xdata.append(x)
        ydata.append(s)
    plt.title("Sine")
    plt.plot(xdata,ydata, "--", color= "g",label =  "dx =0.1")
    plt.xlabel("x")
    plt.ylabel("Area")
    x = np.arange(a,b,0.1)
    plt.legend()
    return

integ(f,0,9,0.1)
```

# 3. (Differentiation) Trigonometric Function
* sin, cos함수의 도함수를 finite difference 방법을 이용하여 구현하고 그래프를 그려라.
* sin, cos함수의 도함수를 scipy.misc.derivative 함수를 이용하여 구하고 그래프를 그려라.

```python

def f(x) : return np.sin(x)
def g(x) : return np.cos(x)

def diff(f,a,b,dx) :
    s,x,k = 0,a,0
    ydata = []
    xdata = np.arange(a,b,dx)
    for i in xdata :
        k = (f(i+dx)/2 - f(i-dx)/2)/dx
        ydata.append(k)
    plt.title('sine')
    plt.plot(xdata,ydata, "-o", color= "g")
    plt.xlabel("x")
    plt.ylabel("Area")
    plt.legend()
    return
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
diff(f,-7,7,0.1)
plt.subplot(1,2,2)
diff(f,-7,7,1)



```
