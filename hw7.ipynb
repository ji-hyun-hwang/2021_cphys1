```python
import numpy as np
import matplotlib.pyplot as plt
```

# Pauli Matrices

```python
Sx = np.array([[0,1],[1,0]])
Sy = np.array([[0,-1j],[1j,0]])
Sz = np.array([[1,0],[0,-1]])

v,s = np.linalg.eig(Sx)
v,s
print("Sx의 고유값이 %s, 고유벡터는 %s,%s "%(v[0],s[0][0],s[1][0]))
print("Sx의 고유값이 %s, 고유벡터는 %s,%s "%(v[1],s[0][1],s[1][1]))

v,s = np.linalg.eig(Sy)
print("Sy의 고유값이 %s, 고유벡터는 %s,%s "%(v[0],s[0][0],s[1][0]))
print("Sy의 고유값이 %s, 고유벡터는 %s,%s "%(v[1],s[0][1],s[1][1]))

v,s = np.linalg.eig(Sz)
print("Sz의 고유값이 %s, 고유벡터는 %s,%s "%(v[0],s[0][0],s[1][0]))
print("Sz의 고유값이 %s, 고유벡터는 %s,%s "%(v[1],s[0][1],s[1][1]))
```

# Normal Modes

```python
m = np.array([[2,-1],[-1,2]])

v,s = np.linalg.eig(m)
print("normal mode의 고유값이 %s, 고유벡터는 %s,%s "%(v[0],s[0][0],s[1][0]))
print("normal mode의 고유값이 %s, 고유벡터는 %s,%s "%(v[1],s[0][1],s[1][1]))
```

# 의미

$$ w_1 = \sqrt{\frac{k}{m}}, \quad w_2 = \sqrt{\frac{3k}{m}}. $$

고요벡터는 위 진동에서 위상을 나타낸다.

$ w_1 $ 은 입자의 진동은 같은 위상임을 나타내고, 입자의 속도도 같음을 의미한다.

$ w_2 $ 은 입자의 진동이 역 위상임을 나타내고, 입자의 속도는 같고 방향은 반대이다.

고유 값은 진동수에 영향을 미치게 되는데, 일반적으로 역 위상인 경우 진동수가 높고, 같은 위상인 경우 대칭 진동수가 낮다.

그러므로 계산된 Normal Mode의 위상과 진동수는 합리적이다.
